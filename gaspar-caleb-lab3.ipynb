```python
locations = {
    "admu":{
        "name":"Ateneo de Manila University",
        "latitude":14.6394557,
        "longitude":121.0758975        
    },
    "upd":{
        "name":"University of the Philippines-Diliman Campus",
        "latitude":14.6538519,
        "longitude":121.0663348
    },
    "dlsu":{
        "name":"De La Salle University-Taft",
        "latitude":14.5647346,
        "longitude":120.9909876

    }, # <-- Note that Python does not complain about this extra comma. This is perfectly okay.
       #     The creators of Python saw that dictionaries tend to be changed a lot, so allowing this is a
       #     convenience for developers who need to add elements when needed.
}

legs = {
     ("upd","admu"):{
         "travel_time_mins":10
     },
     ("admu","dlsu"):{
         "travel_time_mins":35
     },
     ("dlsu","upd"):{
         "travel_time_mins":55
     }
}

def eta(a, b):
    time = 0
    if a != b:
        if (a,b) in legs:
            time = legs[a,b]['travel_time_mins']
        else:            
            # Create New Lists
            newList = list(legs.keys())
            bList = []
            aList = []
            
            for i in newList:
                bList.append(i[1])
                aList.append(i[0])
                
                
            # Calculation
            time = legs[newList[bList.index(b)]]['travel_time_mins'] + legs[newList[aList.index(a)]]['travel_time_mins'] 
    else:
        return 0;
            
    return time

print(eta('admu','upd'))
print(eta('dlsu','admu'))
```
